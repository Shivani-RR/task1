```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'D:\14 Interns\hospital_patients_dirty.csv')
df
df.head()
```

```python
# Identify and handle missing values using .isnull() in Python or filters in Excel.
df.isnull().sum()
```

```python
# Remove duplicate rows using .drop_duplicates() or Excel’s “Remove Duplicates”
df.drop_duplicates(inplace=True)
df
```

```python
# Convert date formats to a consistent type (e.g., dd-mm-yyyy)
df["AdmissionDate"] = pd.to_datetime(df["AdmissionDate"], errors="coerce")
df
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()  # remove spaces
```

```python
#Check and fix data types (e.g., age should be int, date as datetime)
df["age"] = df["age"].astype("Int64")  # allows null-safe conversion
df
```

```python
# Rename column headers to be clean and uniform (e.g., lowercase, no spaces).
df.to_csv(r"D:\14 Interns\hospital_patients_dirty.csv", index=False)
df
```
